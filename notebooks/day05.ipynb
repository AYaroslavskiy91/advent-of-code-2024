```python
import pandas as pd
import numpy as np
import re
```

```python
with open('../input/day05.txt', 'r') as f:
    lines = f.readlines()
lines[1175]
```

```python
order, pages = ''.join(lines).split('\n\n')
```

```python
order_dict = {}
for rule in order.split('\n'):
    key, value = rule.split('|')
    order_dict.setdefault(key, []).append(value)
order_dict
```

```python
pages_list = pages.split('\n')
pages_list[:10]
```

```python
def get_middle(lst):
    len_list = len(lst)
    return int(lst[int((len_list-1)/2)])
```

```python
def check_valid(nums):
    return np.mean([
        ele not in set(nums[:idx])
        for idx 
        in range(1,len(nums))
        for ele in order_dict.get(nums[idx]) 
    ])==1
```

```python
def sort_line(nums):

    for idx in range(1,len(nums)):
        for ele in order_dict.get(nums[idx]):
            if ele in (nums[:idx]):
                idx_old = nums.index(ele)
                srt = nums[:idx_old] + [nums[idx]] + nums[idx_old+1:idx] + [nums[idx_old]] + nums[idx+1:]
        
    if not check_valid(srt):
        srt = sort_line(srt)

    return srt
```

```python
p1 = 0
p2 = 0
for page in pages_list:
    nums = page.split(',')
    if check_valid(nums):
        p1 += get_middle(nums)
    else:
        p2 += get_middle(sort_line(nums))
p1, p2
```
