```python
import pandas as pd
import numpy as np
import re
```

```python
with open('../input/day04.txt', 'r') as f:
    lines = f.readlines()
lines[:2]
```

```python
def checkline(line):
    return len(re.findall('XMAS', line)) + len(re.findall('SAMX', line))
```

```python
def gethvs(df):
    return df.apply(lambda x: ''.join(x), axis=1).apply(checkline).sum() 
```

```python
def getdiag(diag):
    return checkline(''.join(diag))
```

```python
mat = pd.DataFrame([[char for char in line.strip()] for line in lines])
mat
```

```python
hor_ver = np.sum(gethvs(mat) + gethvs(mat.T))
hor_ver
```

```python
diag_sum = 0
for k in range(-len(mat), len(mat)):
    diag_sum += getdiag(np.diag(mat, k))
    diag_sum += getdiag(np.diag(np.fliplr(mat), k))

diag_sum
```

```python
p1 = diag_sum+hor_ver
p1
```

```python
def get_submatrix(npa, r,c):
    return npa.iloc[r:r+3, c:c+3]
```

```python
def checkmas(subm):
    return (''.join(np.diag(subm))=="SAM" or ''.join(np.diag(subm))=="MAS") and (''.join(np.diag(np.fliplr(subm)))=="SAM" or ''.join(np.diag(np.fliplr(subm)))=="MAS")
```

```python
p2 = 0
for r in range(len(mat)):
    for c in range(len(mat)):
        p2 += checkmas(get_submatrix(mat,r,c)) 
```
