```python
import pandas as pd
import numpy as np
import re
from itertools import cycle
```

```python
with open('../input/day06.txt', 'r') as f:
    lines = f.read().splitlines()
lines
```

```python
for x, line in enumerate(lines):
    start = re.search('[><^v]', line)
    if start:
        y = (start.span()[0])
        x_start, y_start = x, y
x_start, y_start
```

```python
def get_move(x, y, direc):
    if direc == '^':
        return x-1, y
    elif direc == 'v':
        return x+1, y
    elif direc == '<':
        return x, y-1
    elif direc == '>':
        return x, y+1
    else:
        print('Something has gone horribly wrong')
```

```python
def get_rotations():
    direcs = ['^', '>', 'v', '<']
    return cycle(direcs)
```

```python
def check_out_bounds(x, y, path):
    return (x < 0 or x >= len(path)) or (y < 0 or y >= len(path[x]))
    
```

```python
def get_new_coords(x, y, direc):
    [x_try, y_try] = get_move(x, y, direc)
    return (x_try, y_try, direc)
```

```python
def add_x(x, y, path):
    path[x] = (path[x][:y] + 'x' + path[x][y+1:])
    return path[x]
```

```python
def get_inits(lines):
    path = lines.copy()
    direc = lines[x_start][y_start]
    [x, y] = [x_start, y_start]
    return [x, y, direc, path]
```

```python
def get_path(lines):
    [x, y, direc, path] = get_inits(lines)
    rotations = get_rotations()
    direc = next(rotations)
    coords = []
    while True:
        path[x] = add_x(x, y, path)
        [x_try, y_try, direc] = get_new_coords(x, y, direc)

        if check_out_bounds(x_try, y_try, path):
            # print('out')
            break;

        if lines[x_try][y_try] == '#':
            # print('obstacle!')
            direc = next(rotations)
            [x_try, y_try, direc] = get_new_coords(x, y, direc)


        if [x_try, y_try, direc] in coords:
            return -1, path

        coords.append([x_try, y_try, direc])
        [x, y] = [x_try, y_try]
    return coords, path
        
```

```python
coords, path = get_path(lines)
len(coords)
```

```python
p1 = 0
for line in path:
    xs = re.findall('x', line)
    p1 += len(xs)
p1
```

```python
p2_list = []
for idx, coord in enumerate(coords):
    x_obs, y_obs = coord[0], coord[1]
    path_obs = lines.copy()
    path_obs[x_obs] = path_obs[x_obs][:y_obs] + '#' + path_obs[x_obs][y_obs+1:]

    coords_tmp, patj_tmp = get_path(path_obs)
    if coords_tmp == -1:
        # print('loop!')
        p2_list.append((x_obs, y_obs))

    if idx % 500 == 0:
        print(f'{idx/len(coords)} percent done')
p2 = len(set(p2_list))
p2
```
