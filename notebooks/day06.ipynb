```python
import pandas as pd
import numpy as np
import re
from itertools import cycle
```

```python
with open('../input/day06.txt', 'r') as f:
    lines = f.read().splitlines()
lines[:5]
```

```python
for x, line in enumerate(lines):
    start = re.search('[><^v]', line)
    if start:
        y = (start.span()[0])
        x_start, y_start = x, y
x_start, y_start
```

```python
lines[x_start][y_start]
```

```python
def get_move(x, y, direc):
    if direc == '^':
        return x-1, y
    elif direc == 'v':
        return x+1, y
    elif direc == '<':
        return x, y-1
    elif direc == '>':
        return x, y+1
    else:
        print('Something has gone horribly wrong')
```

```python
rotations = cycle(['^', '>', 'v', '<'])
```

```python
path = lines.copy()
direc = lines[x_start][y_start]
x, y = x_start, y_start
while True:
    path[x] = (path[x][:y] + 'x' + path[x][y+1:])
    new_x, new_y = get_move(x, y, direc)
    if path[new_x][new_y] == '#':
        direc = next(rotations)
        new_x, new_y = get_move(x, y, direc)
    if (new_x < 0 or new_x > len(path)) or (new_y < 0 or new_y > len(path[x])):
        break
    x, y = new_x, new_y

path
```

```python
total_stops = 0
for line in path:
    xs = re.findall('x', line)
    total_stops += len(xs)
total_stops
```
