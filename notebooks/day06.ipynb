```python
def get_move(x, y, direc):
    if direc == '^':
        return x-1, y
    elif direc == 'v':
        return x+1, y
    elif direc == '<':
        return x, y-1
    elif direc == '>':
        return x, y+1
    else:
        print('Something has gone horribly wrong')
```
