```python
import pandas as pd
import numpy as np
import re
from functools import cache
```

```python
with open('../input/day19.txt', 'r') as f:
    lines = f.read().splitlines()
lines[:5]
```

```python
patterns = re.findall('[a-z]+', lines[0])
patterns
```

```python
def check_string(string):
    if string in patterns:
        return True

    gps = []
    for p in patterns:
        if string.startswith(p):
            gps.append(p)
    return any(check_string(string.replace(p, '')) for p in gps)
```

```python
valid = 0
for line in lines[2:]:
    if check_string(line):
        valid += 1
```

```python
valid
```

```python
@cache
def count_string(string):
    if string == '':
        return 1

    count = 0
    for p in patterns:
        if string.startswith(p):
            # count += count_string(string.replace(p, ''))
            count += count_string(string[len(p):])
    return count
```

```python
%%time
designs = 0
for line in lines[2:]:
    designs += count_string(line)
```
